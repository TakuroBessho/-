```python
##①ライブラリーのインストール

import requests
import pandas as pd
import json


url=('https://api.coingecko.com/api/v3/coins/')+str('bitcoin')+('/market_chart?vs_currency=jpy&days=max')
r=requests.get(url)
r2=json.loads(r.text)
btc=r2['prices']

url3=('https://api.coingecko.com/api/v3/coins/')+str('ethereum')+('/market_chart?vs_currency=jpy&days=max')
r5=requests.get(url3)
r6=json.loads(r5.text)
eth=r6['prices']

url4=('https://api.coingecko.com/api/v3/coins/')+str('bitcoin-cash')+('/market_chart?vs_currency=jpy&days=max')
r7=requests.get(url4)
r8=json.loads(r7.text)
bch=r8['prices']


```

```python
#＃＃＃＃＃＃＃＃＃##ここから始める#＃＃＃＃＃


import numpy as np
import pandas as pd
a = pd.read_csv("btcethchange.csv",index_col=0)
a

#欠損地の削除
b=a.dropna()
b
```

```python
#インデックスの生成
from datetime import datetime
bitcoin_date=[]
for i in b['date']:
    tsdate=int(i)
    loc =datetime.utcfromtimestamp(tsdate)
    bitcoin_date.append(loc)
    
bitcoin_date
```

```python
b.index = pd.to_datetime(bitcoin_date)
b.index
```

```python
import matplotlib.pyplot as plt
from pylab import rcParams
import matplotlib as mpl
font = {"family":"Noto Sans CJK JP"}
mpl.rc('font', **font)
%matplotlib inline
rcParams['figure.figsize'] = 10,5

plt.title('変化率の比較')
plt.subplot(2, 1, 1)
plt.plot(b['btc'],'blue',label='ビットコインの収益率',alpha=1,linewidth=0.5)
plt.grid(which='both')
plt.legend()


plt.subplot(2, 1, 2)
plt.plot(b['eth'],'orange',label='イーサリアムの収益率',alpha=0.7,linewidth=0.5)
plt.legend()
plt.grid(which='both')
```

```python
rcParams['figure.figsize'] = 10,5
plt.plot(b['btc'],'blue',label='ビットコインの収益率',alpha=1,linewidth=0.5)
plt.grid(which='both')
plt.title('ビットコインの収益率の推移')
plt.ylabel('(％)')
plt.legend()
```

```python
##累積変化率の計算&プロット
rcParams['figure.figsize'] = 10,5

ethreturns=(b['eth']+1).copoud()

plt.subplot(2, 1, 2)
plt.plot(b['bitcoin'],'orange',label='イーサリアムの収益率',alpha=0.7,linewidth=0.5)
plt.legend()
plt.grid(which='both')
btcreturns=(b['btc']+1).cumprod()
plt.title('累積収益率の推移と比較')
plt.plot(ethreturns*100,'orange',label='イーサリアムの累積収益率(%)',alpha=0.9,linewidth=0.8)
plt.plot(btcreturns*100,'blue',label='ビットコインの累積収益率(%)',alpha=0.9,linewidth=0.8)
plt.legend()
plt.grid(which='both')
```
